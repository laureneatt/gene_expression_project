# 🧬 DNA (Deoxyribonucleic Acid)

DNA stores all the genetic information of an organism.

It acts as the **instruction manual** for every cell.

### Key Points

- Contains all genes.
- Nearly every cell has the same DNA.
- Only a subset of genes is used in each cell.

---

# 🌍 Genome

The **genome** is the complete set of DNA in an organism.

The human genome contains approximately **20,000 protein-coding genes**.

---

# 📄 Gene

A gene is a segment of DNA.

Each gene contains the instructions needed to produce a protein.

```text
DNA
│
├── Gene A
├── Gene B
├── Gene C
└── ...
```

---

# 🧫 Cell

A cell is the basic unit of life.

Although cells share the same DNA, they do **not** express the same genes.

Examples:

- Brain cell
- Liver cell
- Muscle cell
- Cancer cell

Different gene expression leads to different cell functions.

---

# 🔬 Central Dogma

```text
DNA
 │
 │ Transcription
 ▼
RNA
 │
 │ Translation
 ▼
Protein
```

- DNA stores genetic information.
- RNA carries genetic information.
- Proteins perform biological functions.

---

# 🧬 RNA (Ribonucleic Acid)

RNA is a temporary copy of an active gene.

The more active a gene is,

➡️ the more RNA it produces.

---

# 🧬 Codon

A **codon** is a sequence of three nucleotides.

Each codon specifies one amino acid during protein synthesis.

Example:

```text
ATG GAA TTT
```

---

# 🧪 Protein

Proteins perform most biological functions.

Examples:

- Enzymes
- Antibodies
- Hormones
- Cell receptors
- Structural proteins

Different cells produce different proteins.

---

# 📈 Gene Expression

Gene expression measures **how active a gene is**.

Example:

| Gene | Expression |
|------|-----------:|
| Gene A | 12 |
| Gene B | 350 |
| Gene C | 0 |

Interpretation:

- High value → highly expressed
- Low value → weakly expressed
- 0 → inactive

---

# 🔬 RNA Sequencing (RNA-seq)

RNA sequencing measures the amount of RNA produced by each gene.

It provides the expression level of thousands of genes simultaneously.

In our dataset:

- One column = one gene
- One value = gene expression level

---

# 👤 Sample

A sample is a biological specimen collected from a patient.

In our dataset:

**1 row = 1 sample = 1 patient**

Each patient has a unique gene expression profile.

---

# 🧬 Gene Expression Profile

A gene expression profile is the complete set of expression values measured for one patient.

Example:

```text
Gene 1 = 12
Gene 2 = 315
Gene 3 = 0
...
Gene 20531 = 41
```

This profile represents the patient's **molecular signature**.

# 🤖 Machine Learning

The model never sees the cancer directly.

It only receives thousands of gene expression values.

```text
Gene Expression Matrix
        │
        ▼
Machine Learning Model
        │
        ▼
Cancer Prediction
```

The goal is to learn the relationship between gene expression patterns and cancer types.

---

# 🩺 Why Does It Work?

Cancer cells express genes differently from healthy cells.

Some genes become:

- highly expressed;
- weakly expressed;
- completely inactive.

Each cancer therefore has its own **gene expression signature**.

Machine Learning learns to recognize these signatures.
